```python
list_of_genes = list(set(adata.gene))
print(list_of_genes)
```

```python
for gene in list_of_genes:
    gene_exp = adata[adata['gene'] == gene]
    print(f"Now processing {gene}")
    gene_exp.loc[:,["x","y"]].to_csv(f"./csv/{gene}.csv", index=False, header=False)
    with open(f"./csv/{gene}.csv", 'rb') as f_in:
        with gzip.open(f"./csv_gz/{gene}.csv.gz", 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```
